# Stochastic Gradient Descent (SGD)

## What is Gradient Descent?

Gradient Descent is a general-purpose **optimization algorithm** used to find parameter values that minimize a
cost (or loss) function. Imagine standing on a hilly landscape in thick fog: you can't see the valley floor,
but you *can* feel the slope under your feet. The strategy is simple — take a step downhill, reassess, repeat.

In mathematical terms, for a parameter vector $\mathbf{b}$:

$$ \mathbf{b}_{\text{new}} = \mathbf{b}_{\text{old}} - \alpha \cdot \nabla L(\mathbf{b}_{\text{old}}) $$

where $\alpha$ is the **learning rate** and $\nabla L$ is the gradient of the loss.

### Why *Stochastic*?

Classic (batch) gradient descent computes the gradient over the **entire** dataset before updating. This can be
very slow for large datasets. **Stochastic Gradient Descent** updates the parameters after each *individual*
training example, making it much faster (and adding beneficial noise that can help escape local minima).

### Key Hyper-parameters

| Parameter | Role |
|---|---|
| **Learning Rate** ($\alpha$) | Controls step size. Too large → overshooting; too small → slow convergence. |
| **Epochs** | Number of complete passes through the training data. More epochs → more opportunities to converge. |

## Problem Setup

We will fit a simple linear model:

$$y = 0.4 + 0.8 \cdot x$$

Using SGD, we iteratively refine the coefficients $(b_0, b_1)$ via the update rule:

$$ b = b - \text{learning\_rate} \times \text{error} \times x $$

---
## Step 1: Making Predictions

Before we can *train*, we need a function that computes $\hat{y}$ from an input row and the current
coefficients. For a linear model with one feature:

$$\hat{y} = b_0 + b_1 \cdot x_1$$

The implementation below generalizes to multiple features.

```python
#make prediction

def make_prediction(input_row,coefficients):
    out_put_y_hat = coefficients[0]
    for i in range(len(input_row)-1):
        out_put_y_hat += coefficients[i+1] * input_row[i]
    return  out_put_y_hat

test_dataset = [[1,1],[2,3],[4,3],[3,2],[5,5]]
test_coefficients = [0.4,0.8]

for row in test_dataset:
    y_hat = make_prediction(row,test_coefficients)
    print("True Y value = %.3f, Our_Prediction = %.3f"%(row[-1],y_hat))

```

---
## Step 2: Training with SGD

The training loop works as follows:

1. **Loop** over all epochs.
2. Within each epoch, **loop** over every row in the training set.
3. For each row, compute the prediction error and **update every coefficient**.

The update rules are:

$$b_0(t+1) = b_0(t) - \alpha \cdot \text{error}(t)$$

$$b_1(t+1) = b_1(t) - \alpha \cdot \text{error}(t) \cdot x_1(t)$$

where $\text{error} = \hat{y} - y$ (prediction minus expected value).

### Convergence

As training progresses, the sum of squared errors should decrease. If the learning rate is well-chosen,
the coefficients will converge toward the optimal values. Monitoring the error after each epoch is a
practical way to verify convergence.

### Visualizing Coefficient Updates (Conceptual)

Imagine plotting $b_0$ and $b_1$ over epochs:
- Initially they start at 0 (or random values).
- With each epoch, they move closer to the optimal values (0.4 and 0.8 in our example).
- The trajectory may be noisy (hence *stochastic*), but the overall trend is toward convergence.

A large learning rate causes wild oscillations; a small learning rate produces a smooth but slow path.
