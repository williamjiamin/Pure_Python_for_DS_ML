# Odds and Log-Odds (Logit)

In probability and statistics, **odds** and **log-odds** are alternative ways to express the likelihood of an event. They play a central role in **logistic regression**, where the model predicts the log-odds of the positive class as a linear function of the features.

This notebook covers:
1. Probability review
2. Odds: definition and examples
3. Log-odds (logit function)
4. The connection to logistic regression
5. Python code to compute and visualize these quantities

## 1. Probability Review

A probability $p$ is a number between 0 and 1 representing the chance an event occurs.

| $p$ | Interpretation |
|-----|----------------|
| 0.0 | Impossible |
| 0.5 | Even chance |
| 1.0 | Certain |

While probabilities are intuitive for humans, they have an inconvenient property for modeling: they are bounded between 0 and 1. Linear models produce outputs from $-\infty$ to $+\infty$, so we need a transformation.

## 2. Odds

The **odds** of an event with probability $p$ are defined as:

$$\text{odds} = \frac{p}{1-p}$$

This is the ratio of the probability that the event *occurs* to the probability it *does not*.

**Examples:**
- If $p = 0.5$: odds = $\frac{0.5}{0.5} = 1$ ("even odds")
- If $p = 0.8$: odds = $\frac{0.8}{0.2} = 4$ ("4 to 1 in favor")
- If $p = 0.2$: odds = $\frac{0.2}{0.8} = 0.25$ ("4 to 1 against")

Odds range from $0$ (impossible) to $+\infty$ (certain). They break the upper bound of 1 but are still non-negative.

```python
def odds(p):
    """Compute odds from probability."""
    return p / (1 - p)

probabilities = [0.1, 0.2, 0.3, 0.5, 0.7, 0.8, 0.9, 0.95, 0.99]

print(f"{'Probability':>12}  {'Odds':>10}")
print("-" * 25)
for p in probabilities:
    print(f"{p:>12.2f}  {odds(p):>10.4f}")
```

## 3. Log-Odds (Logit)

The **log-odds** (or **logit**) applies the natural logarithm to the odds:

$$\text{logit}(p) = \ln\left(\frac{p}{1-p}\right)$$

This transformation maps the bounded probability range $(0, 1)$ to the entire real line $(-\infty, +\infty)$:

| $p$ | Odds | Log-odds |
|-----|------|----------|
| 0.01 | 0.0101 | -4.60 |
| 0.5 | 1.0 | 0.0 |
| 0.99 | 99.0 | 4.60 |

**Key insight:** By using log-odds, we can model a non-linear relationship with a linear model. This is exactly how logistic regression works.

```python
from math import log

def logit(p):
    """Compute log-odds from probability."""
    return log(p / (1 - p))

print(f"{'Probability':>12}  {'Odds':>10}  {'Log-Odds':>10}")
print("-" * 37)
for p in probabilities:
    print(f"{p:>12.2f}  {odds(p):>10.4f}  {logit(p):>10.4f}")
```

We can use log-odds to turn the non-intuitive odds into a linear, intuitive model. The figure below shows how odds and probability relate:

## 4. Visualizing the Logit Function

Let's plot the logit function to see how it maps probabilities to the real line.

```python
import math

key_points = [0.01, 0.1, 0.25, 0.5, 0.75, 0.9, 0.99]
print(f"{'p':>6}  {'logit(p)':>10}  Visualization")
print("-" * 50)
for p in key_points:
    lo = math.log(p / (1 - p))
    pos = int(20 + lo * 3)
    pos = max(0, min(40, pos))
    bar = '.' * pos + '|' + '.' * (40 - pos)
    print(f"{p:>6.2f}  {lo:>10.4f}  {bar}")
```

## 5. Connection to Logistic Regression

In logistic regression, we model the **log-odds** of the positive class as a linear function of features:

$$\text{logit}(p) = \ln\left(\frac{p}{1-p}\right) = \beta_0 + \beta_1 x_1 + \beta_2 x_2 + \cdots + \beta_k x_k$$

Solving for $p$, we get the **sigmoid (logistic) function**:

$$p = \frac{1}{1 + e^{-(\beta_0 + \beta_1 x_1 + \cdots)}}$$

This is why the logit transformation is so important: it is the *link function* that connects the linear predictor to the probability output.

```python
import math

def sigmoid(z):
    """Inverse of the logit function."""
    return 1 / (1 + math.exp(-z))

test_probs = [0.1, 0.3, 0.5, 0.7, 0.9]
print(f"{'Original p':>12}  {'logit(p)':>10}  {'sigmoid(logit(p))':>18}")
print("-" * 45)
for p in test_probs:
    lo = logit(p)
    recovered = sigmoid(lo)
    print(f"{p:>12.4f}  {lo:>10.4f}  {recovered:>18.4f}")
```

## 6. Worked Example

Suppose a logistic regression model predicts that a patient has a 75% chance of having diabetes.
Let's compute the odds and log-odds.

```python
p_diabetes = 0.75

odds_diabetes = odds(p_diabetes)
logit_diabetes = logit(p_diabetes)

print(f"Probability of diabetes: {p_diabetes}")
print(f"Odds of diabetes:        {odds_diabetes:.4f}  (3 to 1 in favor)")
print(f"Log-odds of diabetes:    {logit_diabetes:.4f}")
print()
print("Interpretation:")
print(f"  The patient is {odds_diabetes:.0f}x more likely to have diabetes than not.")
print(f"  A logistic regression would output {logit_diabetes:.4f} as the raw score.")
```

## Summary

| Concept | Formula | Range |
|---------|---------|-------|
| Probability | $p$ | $[0, 1]$ |
| Odds | $\frac{p}{1-p}$ | $[0, +\infty)$ |
| Log-Odds (Logit) | $\ln\left(\frac{p}{1-p}\right)$ | $(-\infty, +\infty)$ |
| Sigmoid (inverse logit) | $\frac{1}{1+e^{-z}}$ | $(0, 1)$ |

The logit function bridges bounded probabilities and unbounded linear models, making it the cornerstone of logistic regression.
